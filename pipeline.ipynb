```python
def print_case_res(final_state_of_case):
    if hasattr(final_state_of_case, 'model_dump'):
        state = final_state_of_case.model_dump()
    elif isinstance(final_state_of_case, dict):
        state = final_state_of_case
    else:
        try:
            state = final_state_of_case.model_dump()
        except Exception:
            state = {}

    print(state.get('defendants'))
    print('='*70)
    print(state.get('defense_documents'))
    print('='*70)
    print(state.get('errors'))
    print('='*70)
    print(state.get('confidence_score'))
    print('='*70)
    judge_out = state.get('agent_outputs', {}).get('judge', {})
    for i, j in judge_out.items():
        print(f"{i} :")
        print(j)
        print('-'*50)
```

```python
print_case_res(final)
```

```python
# Quick error diagnostics
state = final.model_dump() if hasattr(final, 'model_dump') else final
print("🔍 ERRORS IN PIPELINE:\n")
errors = state.get('errors', [])
if errors:
    for error in errors:
        print(f"  • {error}")
else:
    print("  ✅ No errors")

print(f"\n📈 AGENTS COMPLETED: {state.get('completed_agents', [])}")
print(f"📊 EXTRACTED DATA COUNTS:")
for agent_name, output in state.get('agent_outputs', {}).items():
    if isinstance(output, dict) and 'extracted_counts' in output:
        print(f"   {output['extracted_counts']}")
```
