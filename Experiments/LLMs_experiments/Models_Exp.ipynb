# LLM Mapping

| Agent | Assigned Model | Rationale |
|------|----------------|-----------|
| **Orchestrator** | No LLM<br>or **GPT-OSS-20M** | Pure logic, routing, and state management |
| **Data Ingestion & Graph Builder Agent** | **GPT-OSS-120M** + **Qwen/Qwen3-32B** | GPT-OSS: text → JSON / triples<br>Qwen: Arabic legal understanding |
| **Procedural Auditor Agent** | **Qwen/Qwen3-32B** | High accuracy in Arabic procedural legality and nullity detection |
| **Legal Researcher Agent** | **Qwen/Qwen3-32B** | Linking charges to legal articles and cassation principles |
| **Evidence Analyst (Parsing)** | **GPT-OSS-120M** | Extraction of testimonies, exhibits, and factual relations |
| **Evidence Analyst (Weighing)** | **LLaMA-3.3-70B-Versatile** | Weighing proof vs. denial (thubut vs. nafi) |
| **Defense Analysis Agent** | **LLaMA-3.3-70B-Versatile** | Construction of complex defense arguments (cassation-style) |
| **Judge Agent (Decision Agent)** | **LLaMA-3.3-70B-Versatile only** | Judicial reasoning, balancing prosecution vs. defense, formal legal reasoning |


# `Note` : Important Constraints

> **Strict Constraints**
> - GPT-OSS models must not perform legal reasoning or judgment
> - No agent may issue a final conclusion except the Judge Agent
> - The Judge Agent consumes only Neo4j subgraphs (Graph RAG output)
> - Any missing or insufficient data must be explicitly declared

# Summary

- GPT-OSS (20M / 120M) → structuring, parsing, validation
- Qwen 32B → Arabic legal comprehension
- LLaMA 70B → deep judicial reasoning and cassation-style justification

## Load Env and Import Lib

```python
from langchain_groq import ChatGroq

import os
from dotenv import load_dotenv
from IPython.display import display , Markdown
load_dotenv()
```

```python
os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')
```

```python
def dis(text):
    print(display(Markdown(text)))
```

## GPT-OSS

```python
gpt_llm = ChatGroq(model="openai/gpt-oss-120b",temperature=0)
```

```python
res1 = gpt_llm.invoke("Hello")
dis(res1.content)
```

## Qwen 32B

```python
qwn_llm = ChatGroq(model="Qwen/Qwen3-32B",temperature=0)
```

```python
res2 = qwn_llm.invoke("Hello")
dis(res2.content)
```
